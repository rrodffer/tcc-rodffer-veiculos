```python
import pandas as pd
```

```python
dataset2023 = pd.read_csv('data/acidentes2023.csv', encoding='ISO-8859-1', delimiter=';')
dataset2022 = pd.read_csv('data/acidentes2022.csv', encoding='ISO-8859-1', delimiter=';')

dataset = pd.concat([dataset2023, dataset2022], ignore_index=True)
```

```python
# limpeza do dataset
dataset.replace("Não Informado", pd.NA, inplace=True)
dataset.replace("Ignorado", pd.NA, inplace=True)
dataset.replace("", pd.NA, inplace=True)
dataset.replace(" ", pd.NA, inplace=True)

dataset = dataset.dropna()

dataset
```
